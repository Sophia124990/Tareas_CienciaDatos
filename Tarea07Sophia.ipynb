Inciso 1


```python
#Generamos un DataFrame con los datos de los cuatro archivos
import pandas as pd

#Leemos los archivos CSV
df1 = pd.read_csv('/content/emisiones-2016.csv', header=0, sep=';', decimal='.')
df2 = pd.read_csv('/content/emisiones-2017.csv', header=0, sep=';', decimal='.')
df3 = pd.read_csv('/content/emisiones-2018.csv', header=0, sep=';', decimal='.')
df4 = pd.read_csv('/content/emisiones-2019.csv', header=0, sep=';', decimal='.')
#Combinamos los DataFrames en uno solo
#Como los archivos tienen las mismas columnas podemos concatenarlos
df_concatenado = pd.concat([df1, df2, df3, df4], ignore_index=True)

#Mostramos las primeras 5 filas para comprobar que se hayan cargado bien los datos
df_concatenado.head()
```

Inciso 2

```python
# Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÃ‘O, MES y las correspondientes a los dÃ­as D01, D02, etc.
columnas_deseadas = ['ESTACION', 'MAGNITUD', 'ANO', 'MES'] + [col for col in df_concatenado.columns if col.startswith('D')]
df_filtrado = df_concatenado[columnas_deseadas]

# Mostramos las primeras filas del DataFrame filtrado
df_filtrado.head()
```

Inciso 3

```python
#Reestructuramos el DataFrame para que los valores de los contaminantes de las columnas de los dÃ­as aparezcan en una Ãºnica columna
df_reestructurado = pd.melt(df_filtrado, id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], value_vars=[col for col in df_filtrado.columns if col.startswith('D')], var_name='DIA', value_name='VALOR')
#Mostramos las primeras 5 filas para comprobar la informaciÃ³n
df_reestructurado.head()
```

Inciso 4


```python
#AÃ±adir una columna con la fecha a partir de la concatenaciÃ³n del aÃ±o, el mes y el dÃ­a (usar el mÃ³dulo datetime)
import datetime

def crea_fecha(row):
    anno = str(row.ANO)
    mes = str(row.MES).zfill(2)

    #Como 'DIA' tiene el formato 'D01', 'D02', etc., extraemos solo el nÃºmero
    dia = str(row.DIA)[1:].zfill(2)

    #Creaamos la fecha manejando los posibles errores como dia 30 con mes febrero
    try:
        fecha = datetime.date(int(anno), int(mes), int(dia))
    except ValueError:
        fecha = None  # Asignar None si la fecha no es vÃ¡lida
    return fecha

#Aplicamos la funciÃ³n para crear la columna 'FECHA'
df_reestructurado['FECHA'] = df_reestructurado.apply(crea_fecha, axis=1)

#Mostramos las primeras 5 filas para comprobar la informaciÃ³n
df_reestructurado.head()
```

```python
#Crear una funciÃ³n que devuelva las emisiones medias mensuales de un contaminante y un aÃ±o dados para todas las estaciones
```

```python
import pandas as pd

def emisiones_medias_mensuales(df, contaminante, aÃ±o):
    # Convertir la columna 'FECHA' a formato datetime si no lo estÃ¡
    if df['FECHA'].dtype != 'datetime64[ns]':
        df['FECHA'] = pd.to_datetime(df['FECHA'])

    # Filtrar por contaminante y aÃ±o
    df_filtrado = df[(df['MAGNITUD'] == contaminante) & (df['FECHA'].dt.year == aÃ±o)]

    # AÃ±adir columna de mes para agrupar
    df_filtrado['MES'] = df_filtrado['FECHA'].dt.month

    # Agrupar por estaciÃ³n y mes y calcular la media de las emisiones
    emisiones_medias = df_filtrado.groupby(['ESTACION', 'MES'])['VALOR'].mean().reset_index()

    # Renombrar la columna de la media para mayor claridad
    emisiones_medias.rename(columns={'VALOR': 'EMISION_MEDIA_MENSUAL'}, inplace=True)

    # Imprimir las primeras 5 filas del resultado
    print(emisiones_medias.head())

    return emisiones_medias
```

```python
#Crear una funciÃ³n que reciba una estaciÃ³n de mediciÃ³n y devuelva un DataFrame con las medias mensuales de los distintos tipos deÂ contaminante
import pandas as pd

def medias_mensuales_por_contaminante(df, estacion):
    # Convertir la columna 'FECHA' a formato datetime si no lo estÃ¡
    if df['FECHA'].dtype != 'datetime64[ns]':
        df['FECHA'] = pd.to_datetime(df['FECHA'])

    # Filtrar los datos por la estaciÃ³n de mediciÃ³n
    df_filtrado = df[df['ESTACION'] == estacion]

    # AÃ±adir columna de mes para agrupar
    df_filtrado['MES'] = df_filtrado['FECHA'].dt.month

    # Agrupar por contaminante y mes, y calcular la media de las emisiones
    medias_mensuales = df_filtrado.groupby(['MAGNITUD', 'MES'])['VALOR'].mean().reset_index()

    # Renombrar la columna de la media para mayor claridad
    medias_mensuales.rename(columns={'VALOR': 'EMISION_MEDIA_MENSUAL'}, inplace=True)

    # Imprimir las primeras 5 filas del resultado
    print(medias_mensuales.head())

    return medias_mensuales
```
